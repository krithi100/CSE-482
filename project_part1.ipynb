```python
from pymongo import MongoClient
import pandas as pd
```

```python
client = MongoClient('localhost', 27017)

db = client['nfl_db']
collection = db.nfl_coll

collection.drop()
```

```python
df = pd.read_csv('nfl_data.csv')
table_names = df['TableName']
df.loc[table_names == 'Team']
```

```python
nfl_dict = df.to_dict('record')
collection.insert_many(nfl_dict)
```

```python
collection.count_documents({})
```

```python
for record in collection.find({"TableName" : "Team"}):
    print(record)
```

```python
for record in collection.find():
    print(record)
```
